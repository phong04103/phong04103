```python
def merge(L, n1, R, n2, a, n):
    i = j = k = 0

    while (i < n1 and j < n2):
        if L[i] < R[j]:
            a[k] = L[i]
            i += 1
        else:
            a[k] = R[j]
            j += 1
        k += 1
    
    while (i < n1):
        a[k] = L[i]
        i += 1
        k += 1
    while (j < n2):
        a[k] = R[j]
        j += 1
        k += 1

def mergeSort(a,n):
    if n > 1:
        n1 = n // 2
        n2 = n - n1
        L = a[:n1]
        R = a[n1:]

        mergeSort(L, n1)
        mergeSort(R, n2)
        merge(L, n1, R, n2,a, n)

a = [5,2,3,1,7]
mergeSort(a,len(a))
a
```

# 11. Student Sort


Cho N sinh viên khác nhau, mỗi sinh viên sẽ gồm mã số sinh viên và điểm của mình. Hãy tìm k sinh viên có điểm cao nhất trong N sinh viên này.

**Dữ liệu nhập**

Dòng thứ nhất gồm N (1≤N≤1000)(1≤N≤1000) là số lượng sinh viên và k (1≤k≤N)(1≤k≤N) là số sinh viên cần lấy điểm cao nhất cách nhau bởi một khoảng trắng.

NN dòng tiếp theo, mỗi dòng gồm mã số sinh viên và điểm của sinh viên đó, điểm của sinh viên là một số thực xx (0≤x≤10)(0≤x≤10).

**Dữ liệu xuất**

Gồm k dòng, mỗi dòng là một mã số của k sinh viên có điểm cao nhất.

Khi hai sinh viên có điểm bằng nhau thì ưu tiên sinh viên có mã số sinh viên nhỏ hơn.

Được biết rằng: Mã số của các sinh viên khác nhau là khác nhau.

Ví dụ
- input <br>

4 3 <br>
15478 8.9 <br>
17895 6.8 <br>
18647 6.8 <br>
14782 9.0 <br>

- output <br>

14782 <br>
15478 <br>
17895 <br>

<hr>Note: x = Student(int(maso),float(diem)) <hr>

```python
class Student:
    def __init__(self,maso,diem):
        self.maso = maso
        self.diem = diem

    def compare(self,orther):
        return self.diem > orther.diem or (self.diem == orther.diem and self.maso < orther.maso)

def merge(L, n1, R, n2, a):
    i = j = k = 0

    while (i < n1 and j < n2):
        if L[i].compare(R[j]):
            a[k] = L[i]
            i += 1
        else:
            a[k] = R[j]
            j += 1
        k += 1
    
    while (i < n1):
        a[k] = L[i]
        i += 1
        k += 1
    while (j < n2):
        a[k] = R[j]
        j += 1
        k += 1

def mergeSort(a,n):
    if n > 1:
        n1 = n // 2
        n2 = n - n1
        L = a[:n1]
        R = a[n1:]

        mergeSort(L, n1)
        mergeSort(R, n2)
        merge(L, n1, R, n2,a)
        
n, k = map(int, input().split(' '))

lst = []

for _ in range(n):
    maso, diem = input().split(' ')
    x = Student(int(maso),float(diem))
    lst.append(x)

mergeSort(lst,n)

for i in range(0,k):
    print(f'{lst[i].maso}')
```

Node đầu tiên chứa x.
Node thứ 2 chứa y.
Các node sau, mỗi node chứa tổng của 2 node liền trước nó (lấy phần dư khi chia cho 1.000.007).

Biết rằng danh sách này có n node.

**Dữ liệu nhập**

Dữ liệu đầu vào chứa một dòng duy nhất chứa 3 số nguyên x, y và n. (0≤x,y≤106,2≤n≤105)(0≤x,y≤10​6​​,2≤n≤10​5​​).

**Dữ liệu xuất**

Một dòng duy nhất, in ra nn số nguyên là các phần tử trong danh sách liên kết, các số ngăn cách nhau bởi một khoảng trắng.
Ví dụ:

input

1 1 5

output

1 1 2 3 5 

Giải thích ví dụ

Node thứ 1 chứa x nên bằng 1.

Node thứ 2 chứa y nên bằng 1.

Từ node thứ 3 trở đi, node sau bằng tổng của 2 node liền trước nó.

Nên node thứ 3 = node thứ 2 + node thứ 1 = 1 + 1 = 2.

Node thứ 4 = node thứ 3 + node thứ 2 = 2 + 1 = 3.

Node thứ 5 = node thứ 4 + node thứ 3 = 3 + 2 = 5.

# 14. Running Time Calculation

```python
import time
def fib(idx):
    seq = [0,1]
    for i in range(idx):
        seq.append(seq[-1]+seq[-2])
    return seq[-2]

def Fib(idx):
    if idx <= 1:
        return idx
    else:
        return Fib(idx - 1) + Fib(idx - 2)

print("**** recruision ****")
rec = time.time()
print(Fib(8))
print("speed: " + str(time.time()-rec))

print("***** iteration ******")
it = time.time()
print(fib(8))
print("speed: " + str(time.time()-it))
```

### 2. Double Linked List

```python
class Node:
    def __init__(self,x):
        self.data = x
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0


    def insertAtStart(self,x):
        p = Node(x)
        if (self.head == None):
            self.head = self.tail = p
        else:
            p.next = self.head
            self.head.prev = p
            self.head = p
        self.count += 1

    def insertAtLast(self,x):
        p = Node(x)
        if (self.head == None):
            self.head = self.tail = p
        else:
            p.prev = self.tail
            self.tail.next = p
            self.tail = p
        self.count += 1

    def insertAtIntermediate(self,x,index):
        if (index < 1 or index > self.count):
            print("Invalid index")
            return
        cur = self.head
        prev = self.head
        p = Node(x)
        count = 1

        if index == 1:
            self.insertAtStart(x)
        else:
            while cur:
                if index == count:
                    p.prev = prev
                    p.next = cur
                    prev.next = p
                    cur.prev = p
                
                count += 1
                prev = cur
                cur = cur.next
                self.count += 1

    def delete(self,x):
        cur = self.head
        node_deleted = False

        if cur is None:
            print("List is empty")

        elif cur.data == x:
            #Item to be deleted is found at starting of the list
            self.head.prev = None
            node_deleted = True
            self.head = cur.next

        elif self.tail.data == x:
            #Item to be deleted is found at the end of list
            self.tail = self.tail.prev
            self.tail.next = None
            node_deleted = True
        else:
            while cur:
                #search item to be deleted, and delete that node
                if cur.data == x:
                    cur.prev.next = cur.next
                    cur.next.prev = cur.prev
                    node_deleted = True
                cur = cur.next

        if node_deleted == False: #Item to be deleted is not found in the list
            print("Item not found")
            
        if node_deleted:
            self.count -= 1
                
    def print(self):
        cur = self.head
        while cur:
            print(cur.data, end=" ")
            cur = cur.next

lst = DoublyLinkedList()
while True:
    x = int(input())
    if x == 0:
        break
    lst.insertAtLast(x)

# lst.insertAtStart(6)
lst.insertAtIntermediate(7,5)
lst.delete(6)

lst.print() 
```

```python
class Node:
    def __init__(self,x):
        self.data = x
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def push(self,x):
        p = Node(x)
        if self.top == None:
            self.top = p
        else:
            p.next = self.top
            self.top = p

        self.size += 1

    def pop(self):
        if self.top == None: 
            print("Stack is empty")
        else:
            if self.top.next: #check if there is more than one node
                self.top = self.top.next
            else:
                self.top = None
        self.size -= 1

    def pop(self): #book reference
        if self.top:
            data = self.top.data
            self.size -= 1
            if self.top.next: #check if there is more than one node
                self.top = self.top.next
            else:
                self.top = None
            return data
        else:
            print("Stack is empty")

    def peek(self):
        if self.top:
            return self.top.data
        else:
            print("Stack is empty")



words = Stack()
words.push("Egg")
words.push("Ham")
words.push("Spam")
# words.pop()
# words.pop()
# words.pop()
# words.pop()

cur = words.top
while cur:
    print(cur.data)
    cur = cur.next

words.peek()
```

## 4. Binary Search

```python
def binary_search(arr, start, end, key):
    while start <= end:  
        mid = start + (end - start)//2
        if arr[mid] == key:  
            return mid  
        elif arr[mid] < key:  
            start = mid + 1  
        else:  
            end = mid - 1  
    return -1  


arr = [ 4, 6, 9, 13, 14, 18, 21, 24, 38] 
x = 13
result = binary_search(arr, 0, len(arr)-1, x)  
print(result) 
```

```python
def preorder(root_node):
    current = root_node
    if current is None:
        return
    print(current.data)
    preorder(current.left_child)
    preorder(current.right_child)
preorder(n1)
```

```python
# from collections import deque
class Node:
    def __init__(self, data):
        self.data = data
        self.right_child = None
        self.left_child = None

n1 = Node("root node")
n2 = Node("left child node")
n3 = Node("right child node")
n4 = Node("left grandchild node")
n1.left_child = n2
n1.right_child = n3
n2.left_child = n4

def level_order_traversal(root_node):
    list_of_nodes = []
    traversal_queue = deque([root_node])
    while len(traversal_queue) > 0:
        node = traversal_queue.popleft()
        list_of_nodes.append(node.data)
        if node.left_child:
            traversal_queue.append(node.left_child)
            if node.right_child:
                traversal_queue.append(node.right_child)
    return list_of_nodes
    
print(level_order_traversal(n1))
```

Expression trees

```python
class TreeNode:
    def __init__(self, data=None):
        self.data = data
        self.right = None
        self.left = None

class Stack:
    def __init__(self):
        self.elements = []

    def push(self, item):
        self.elements.append(item)

    def pop(self):
        return self.elements.pop()

expr = "4 5 + 5 3 - *".split()
stack = Stack()

for term in expr:
    if term in "+-*/":
        node = TreeNode(term)
        node.right = stack.pop()
        node.left = stack.pop()
    else:
        node = TreeNode(int(term))
    stack.push(node)

def calc(node):
    if node.data == "+":
        return calc(node.left) + calc(node.right)
    elif node.data == "-":
        return calc(node.left) - calc(node.right)
    elif node.data == "*":
        return calc(node.left) * calc(node.right)
    elif node.data == "/":
        return calc(node.left) / calc(node.right)
    else:
        return node.data

root = stack.pop()
result = calc(root)
print(result)
```

# 17. Object Oriented Programming

## 1. Getter, Setter

```python
class Dog:
    def __init__(self, name, age):
        self._name = name
        self.age = age

    def get_name(self):
        return self._name
    
    def set_name(self, new_name):
        if isinstance(new_name, str) and new_name.isalpha():
            self._name = new_name
        else:
            print("Please enter a valid name.")

my_dog = Dog("Nora", 8)

print("My dog is: ", my_dog.get_name())

my_dog.set_name("Norita")

print("Her new name is:", my_dog.get_name())
```

```python
class Circle:

    def __init__(self, radius):
        self._radius = radius

    def get_radius(self):
        return self._radius
    
    def set_radius(self, new_radius):
        if isinstance(new_radius, float) and new_radius > 0:
            self._radius = new_radius
        else:
            print("Please enter a valid value for the radius")

my_circle = Circle(5.0)
print(my_circle.get_radius())

my_circle.set_radius(10.5)
print(my_circle.get_radius())
```

## 2. Property

```python
class Dog:
    def __init__(self,age):
        self._age = age

    def get_age(self):
        print("Calling the getter ...")
        return self._age
    
    def set_age(self, new_age):
        print("Calling the setter...")
        if isinstance(new_age, int) and 0 < new_age < 30:
            self._age = new_age
        else:
            print("Please enter a valid age.")

    age = property(get_age, set_age)

my_dog = Dog(8)

print(f" My dog is {my_dog.age} years old.")
print("One year later...")

my_dog.age += 1

print(f"My dog is now {my_dog.age} years old.")

```

## 3. super().<method_name>(<arguments>)

```python
class Polygon:

    def __init__(self, num_sides, color):
        self.num_sides = num_sides
        self.color = color

    def describe_polygon(self):
        print(f"This polygon has {self.num_sides} sides and it's {self.color}")

class Triangle(Polygon):

    NUM_SIDES = 3

    def __init__(self, base, height, color):
        super().__init__(Triangle.NUM_SIDES, color)
        self.base = base
        self.height = height

    def find_area(self):
        return (self.base * self.height) / 2

class Square(Polygon):
    
    NUM_SIDES = 4

    def __init__(self, side_length, color):
        super().__init__(Square.NUM_SIDES, color)
        self.side_length = side_length

    def find_area(self):
        return self.side_length ** 2

my_triangle = Triangle(5,4, "Blue")
my_triangle.describe_polygon()
print(my_triangle.find_area())

my_square = Square(4, "Green")
my_square.describe_polygon()
print(my_square.find_area())


```
